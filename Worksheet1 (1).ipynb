For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.



```python
a = np.array([[3,4],[7,8]]);
b = np.array([[5,3],[2,1]]);
a_inv=np.linalg.inv(a)
print("This presents the inverse", a_inv)
I= np.dot(a, a_inv)
print("This presents the idenity matrix: ", I)
print( "Result:", np.allclose(I , np.eye(2)))

```

2. Prove AB ̸= BA.

```python
a = np.array([[3,4], [7,8]]);
b = np.array([[5,3], [2,1]]);
ab = a @ b
ba = b @ a
print("AB: ", ab)
print("BA: ", ba)

print("Result:",np.array_equal(ab,ba))
```

3. Prove (AB)

```python
a = np.array([[3,4], [7,8]]);
b = np.array([[5,3], [2,1]]);
ab= a@b
ab_transpose=ab.T
b_transpose=b.T
a_transpose=a.T
b_transpose_a_transpose= b_transpose@a_transpose;
print("The transpose of AB",ab_transpose);
print("The transpose of b and a", b_transpose_a_transpose)
print("The output of AB transpose and B transpose A transpose: ", np.array_equal(ab_transpose, b_transpose_a_transpose))

```

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

```python


# Coefficient matrix
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

# Constant vector
b = np.array([-1, -3, 9])

# Compute inverse of A
A_inv = np.linalg.inv(A)

# Solve for x
x = A_inv @ b

print("Solution vector [x, y, z] =\n", x)
```
