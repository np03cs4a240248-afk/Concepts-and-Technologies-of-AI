```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python

iris = load_iris()  # Load dataset
X, y = iris.data, iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(   # Spliting data
    X, y, test_size=0.3, random_state=42
)
```

```python
knn = KNeighborsClassifier()  #defining knn
```

```python
param_grid = {"n_neighbors": range(1, 21),"weights": ["uniform", "distance"],
    "p": [1, 2]}
```

```python
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring="accuracy", n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_knn = grid_search.best_estimator_
print("=== KNN Hyperparameter Tuning ===")
print("Best parameters:", grid_search.best_params_)
print("Best CV Accuracy:", grid_search.best_score_)
```

```python
y_train_pred = best_knn.predict(X_train)
y_test_pred = best_knn.predict(X_test) # Predictions and accuracy
print("Train Accuracy:", accuracy_score(y_train, y_train_pred))
print("Test  Accuracy:", accuracy_score(y_test, y_test_pred))
```
